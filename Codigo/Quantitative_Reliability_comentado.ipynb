```
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```
from reliability.Distributions import *
from scipy import *
```

```
from reliability.Fitters import Fit_Everything
data = [4.37, 19.49, 1.72, 0.85, 5.59, 11.18, 23.41, 23.02, 10.79, 13.31, 49.62, 7.26, 13.9, 30.6, 3.56, 25.05, 4.64, 3.63, 11.28, 35.71, 56.13, 6.45, 6.61, 17.85, 5.23]
results=Fit_Everything(failures=data, show_histogram_plot=True, show_probability_plot=True, show_PP_plot=True)
print('The best fitting distribution was', results.best_distribution_name, 'which had parameters', )results.best_distribution.parameters
```

```
results.best_distribution.parameters
```

```
print(results.best_distribution_name)
```

```
results.best_distribution.plot()
```

```
results.best_distribution.SF(label='Fitted Distribution',color='steelblue')
plt.legend()
plt.show()
```

```
results.best_distribution.SF(15)
```

```
results.best_distribution.SF([5,15,20,25])
```

```
results.best_distribution.CDF(label='Fitted Distribution',color='steelblue')
plt.legend()
plt.show()
```
